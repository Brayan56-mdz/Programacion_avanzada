```
import math #importamos la libreria math
class cuadrado:
    def __init__(self, a):
        self.ancho=a
        
    def perimetro(self):
        return  4*(self.ancho) 
    def area(self):
        return (self.ancho)**2

class rectangulo:
    def __init__(self, A,B):
        self.Base=B
        self.Altura=A
        
    def perimetro(self):
        return (self.Altura + self.Base)*2 
    def area(self):
        return self.Base *self.Altura

class triangulo:
    def __init__(self, a,b,c,h):
        self.a=a
        self.b=b
        self.c=c
        self.h=h

    def perimetro(self):
        return self.a +self.b+self.c
    def area(self):
        return (self.b*self.h)/2

class circulo:
    def __init__(self, r):
        self.r=r
        
    def perimetro(self):
        return  3.1416*(self.r)*2 
    def area(self):
        return  3.1416*((self.r)**2 )

class rombo:
    def __init__(self, D,d,l):
        self.D=D
        self.d=d
        self.l=l
    def area(self):
        return ((self.D *self.d)/2 )
    def perimetro(self):
        return (self.l * 4)   

class trapecio:
    def __init__(self, B,b,h):
        self.B=B
        self.b=b
        self.h=h
    def area(self):
        return ((self.B +self.b + self.h ))
    def perimetro(self):
        return ((self.B +(self.b* self.h))/2)

print("-Programa que calcula area y perimetro-")
print("1-Cuadrado")
print("2-Rectangulo")
print("3-Triangulo")
print("4-Circulo")
print("5-Rrombo")
print("6-Trapecio")

op=int(input("Elije una opcion: "))
if op == 1:
    print("-Cuadrado-")
    a=int(input("Ingresa el lado: "))
    #instancia  
    obj1=cuadrado(a)
    print("Su perimetro es: ",obj1.perimetro())
    print("Su area es: " ,obj1.area())

if op == 2:
    print("-rectangulo-")
    A=int(input("dame el tamaño de tu base: "))
    B=int(input("dame el tamaño de tu anch0: "))
    #instancia  
    obj2=rectangulo(A,B)
    print("Su perimetro es: ",obj2.perimetro())
    print("Su area es: " ,obj2.area())

if op == 3 :
    print("-Triangulo-")
    b=int(input("Ingresa base: "))
    a=int(input("Ingresa lado1: "))
    c=int(input("Ingresa lado2: "))
    h=int(input("Ingresa altura: "))
    #instancia
    obj3=triangulo(a,b,c,h)
    print("Su perimetro es: ",obj3.perimetro())
    print("Su area es: " ,obj3.area())

if op == 4:
    print("-Circulo-")
    r=int(input("Ingresa el radio: "))
    #instancia  
    obj4=circulo(r)
    print("Su perimetro es: ",obj4.perimetro())
    print("Su area es: " ,obj4.area())

if op == 5:
    print("-rombo-")
    l=int(input("Ingresa el tamaño de sus lados: "))
    D=int(input("Ingresa  el tamaño de su diagonal mayor: "))
    d=int(input("Ingresa  el tamaño de su diagonal menor: "))
    #instancia 
    obj5=rombo(l,D,d)
    print("Su perimetro es: ",obj5.perimetro())
    print("Su area es: " ,obj5.area())

if op == 6:
    print("-trapecio-")
    B=int(input("Ingresa  el tamaño de su base mayor: "))
    b=int(input("Ingresa  el tamaño de su base menor: "))
    h=int(input("Ingresa  el tamaño de su altura : "))
    
    obj6=trapecio(B,b,h)
    print("Su perimetro es: ",obj6.perimetro())
    print("Su area es: " ,obj6.area())
```
